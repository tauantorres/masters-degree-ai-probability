```python
import os
import subprocess

def mac_pptx_to_pdf(input_path, output_path):
    # Ensure paths are absolute
    input_path = os.path.abspath(input_path)
    output_path = os.path.abspath(output_path)

    script = f'''
    tell application "Microsoft PowerPoint"
        open POSIX file "{input_path}"
        set theActivePres to active presentation
        save theActivePres in POSIX file "{output_path}" as save as PDF
        close theActivePres saving no
    end tell
    '''
    
    # Run the AppleScript from Python
    subprocess.run(['osascript', '-e', script])

mac_pptx_to_pdf(
    "Materials/02/10. Conditional_probability.pptx", 
    "Materials/01/output.pdf",
)
```
